```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
p = np.linspace(0.01, 0.99, 200, endpoint=True)

def entropy(p):
    q = 1 - p
    return -(p*np.log2(p) + q*np.log2(q))

h = entropy(p)
plt.plot(p, h)
plt.title("Entropy")
plt.xlabel("p1")
plt.ylabel("H")
plt.ylim(0, 1.3)
plt.show()
```
